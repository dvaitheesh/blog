```python
from faker import Faker
import pandas as pd
```

```python
Faker.seed(72)
fake = Faker()
```

```python
ds = pd.DataFrame(columns = ['fake phone numbers'])
```

```python
for _ in range(15):
    ds = ds.append({'fake phone numbers': fake.phone_number()}, ignore_index=True)
```
